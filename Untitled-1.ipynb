```python
import cv2
import numpy as np
import os
"""Charge les images et les labels à partir des répertoires.

  Args:
      data_dir: Chemin d'accès au répertoire de données.
      labels: Liste des libellés de classe.
      img_size: Taille de redimensionnement des images carrées (hauteur et largeur).

  Returns:
      tuple: Un tuple contenant deux tableaux NumPy - un pour les images et un pour les labels.
  """

def load_images(data_dir, labels, img_size):
  data_images = []
  data_labels = []

  for label in labels:
    path = os.path.join(data_dir, label)
    class_num = labels.index(label)
    for img in os.listdir(path):
      try:
        img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)
        if img_arr is not None:
          resized_arr = cv2.resize(img_arr, (img_size, img_size))
          resized_arr_rgb_format = cv2.cvtColor(resized_arr, cv2.COLOR_BGR2RGB)

          # Vérifier la forme de l'image redimensionnée
          assert resized_arr_rgb_format.shape == (img_size, img_size, 3)

          data_images.append(resized_arr_rgb_format)
          data_labels.append(class_num)
      except Exception as e:
        print(f"Erreur de chargement de l'image {img}: {e}")

  return np.array(data_images), np.array(data_labels)


# Chemins d'accès pour les données
data_dir = 'C:\\Users\\chari\\Desktop\\chest_xray\\chest_xray'
labels = ['PNEUMONIA', 'NORMAL']
img_size = 224

# Chargement des données
train_images, train_labels = load_images(os.path.join(data_dir, 'train'), labels, img_size)
test_images, test_labels = load_images(os.path.join(data_dir, 'test'), labels, img_size)
val_images, val_labels = load_images(os.path.join(data_dir, 'val'), labels, img_size)

# Fusionner les ensembles de données
dataset_images = np.concatenate((train_images, val_images, test_images), axis=0)
dataset_labels = np.concatenate((train_labels, val_labels, test_labels), axis=0)

print("Formes des tableaux d'images et de labels après fusionnement :")
print(f"dataset_images.shape: {dataset_images.shape}")
print(f"dataset_labels.shape: {dataset_labels.shape}")
```

```python
import random
import matplotlib.pyplot as plt

def plot_images_from_folder(dataset):
    """
    This function displays 8 random images and their corresponding labels from the dataset.

    Args:
        dataset: A list containing tuples of (image, label).
    """
    # Generate 8 random indices within the range of your dataset
    random_indices = random.sample(range(len(dataset)), 8)
    
    plt.figure(figsize=[14, 24])
    for i in range(8):
        plt.subplot(6, 4, i+1)
        plt.imshow(dataset[random_indices[i]][0], cmap='gray')  # Display in grayscale
        plt.axis('off')
        plt.title(dataset[random_indices[i]][1])
    
    plt.show()

# Créer un ensemble de données au format approprié
# Cet ensemble de données sera une liste de tuples où chaque tuple contient l'image et son libellé correspondant
dataset = list(zip(dataset_images, dataset_labels))

# Appeler la fonction plot_images_from_folder avec l'ensemble de données
plot_images_from_folder(dataset)
```

```python
from sklearn.model_selection import train_test_split

# Diviser l'ensemble de données en ensembles d'entraînement et de test
initial_train_data, test_data, initial_train_labels, test_labels = train_test_split(dataset_images, dataset_labels, test_size=0.20, random_state=30)

# Créer des ensembles de données au format approprié pour l'entraînement initial et le test
initial_train_dataset = list(zip(initial_train_data, initial_train_labels))
test_dataset = list(zip(test_data, test_labels))

# Diviser l'ensemble de données d'entraînement initial en ensembles d'entraînement et de validation
train_data, val_data, train_labels, val_labels = train_test_split(initial_train_data, initial_train_labels, test_size=0.20, random_state=30)

# Créer des ensembles de données au format approprié pour l'entraînement et la validation
train_dataset = list(zip(train_data, train_labels))
val_dataset = list(zip(val_data, val_labels))
```

```python
import seaborn as sns

def count_labels(dataset):
    """
    This function displays the count for the label for each class.

    Args:
        dataset: A list containing tuples of (image, label).
    """
    extracted_labels = [data[1] for data in dataset]
    print("Number of labels:", len(extracted_labels))
    
    # Count the occurrences of each class label
    class_counts = np.bincount(extracted_labels)

    # Print the counts
    count_0 = class_counts[0]
    count_1 = class_counts[1]
    print("Count of 0:", count_0)
    print("Count of 1:", count_1)
    
    # Plot the count of labels
    sns.set_style('darkgrid')
    sns.countplot(x=extracted_labels).set(title="Training Data")

# Class distribution for training dataset
count_labels(train_dataset)  # Change 'train_df' to 'train_dataset'
```

```python
# Separate the images and labels for the training dataset
x_train = [data[0] for data in train_dataset]
y_train = [data[1] for data in train_dataset]

# Separate the images and labels for the test dataset
x_test = [data[0] for data in test_dataset]
y_test = [data[1] for data in test_dataset]

# Separate the images and labels for the validation dataset
x_val = [data[0] for data in val_dataset]
y_val = [data[1] for data in val_dataset]





```

```python
# Normalize the data
x_train = np.array(x_train) / 255.0
x_val = np.array(x_val) / 255.0
x_test = np.array(x_test) / 255.0
```

```python
# Reshape the training data
x_train = x_train.reshape(-1, img_size, img_size, 3)
y_train = np.array(y_train)

# Reshape the validation data
x_val = x_val.reshape(-1, img_size, img_size, 3)
y_val = np.array(y_val)

# Reshape the test data
x_test = x_test.reshape(-1, img_size, img_size, 3)
y_test = np.array(y_test)
```

```python
# With data augmentation to prevent overfitting and handling the imbalance in dataset

datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # apply ZCA whitening
        rotation_range = 30,  # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range = 0.2, # Randomly zoom image 
        width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
        horizontal_flip = True,  # randomly flip images
        vertical_flip=False)  # randomly flip images


datagen.fit(x_train)
```

```python
# Define the early stopping and learning rate reduction callback
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=5, restore_best_weights=True)
learning_rate_reduction = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', patience = 3)
```

```python
###################################################VGG 16 Mode##########################################
#Loading the model
from tensorflow.keras.applications.vgg16 import VGG16

vgg16_base_model = VGG16(
    include_top=False,
    weights="imagenet",
    input_shape=(224, 224, 3),
)

#Making sure the layers of the VGG16 model are not retrained 
for layer in vgg16_base_model.layers:
    layer.trainable = False
```

```python
vgg16_model = tf.keras.models.Sequential()
vgg16_model.add(vgg16_base_model)
vgg16_model.add(tf.keras.layers.Flatten())
vgg16_model.add(tf.keras.layers.BatchNormalization())
vgg16_model.add(tf.keras.layers.Dense(128,activation='relu'))
vgg16_model.add(tf.keras.layers.Dropout(0.5))
vgg16_model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
vgg16_model.summary()

vgg16_model.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)
```

```python
# Train the vgg16 model with early stopping
vgg16_model_history = vgg16_model.fit(
    datagen.flow(x_train,y_train, batch_size = 32),
    epochs=30,
    validation_data=datagen.flow(x_val, y_val),
    callbacks=[early_stopping, learning_rate_reduction])
```

```python
#Plotting the VGG16 model results

#Getting the accuracy
acc = vgg16_model_history.history['accuracy']
val_acc = vgg16_model_history.history['val_accuracy']

#Getting the losses
loss = vgg16_model_history.history['loss']
val_loss = vgg16_model_history.history['val_loss']

#No of epochs it trained
epochs_range = vgg16_model_history.epoch

#Plotting Training and Validation accuracy
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

#Plotting Training and Validation Loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
###############################Vgg16 Performance Evaluation#######################################
evaluation_result=vgg16_model.evaluate(x_test,y_test)
```

```python
print("Loss of the model is - " , evaluation_result[0])
print("Accuracy of the model is - " , evaluation_result[1]*100 , "%")
```
